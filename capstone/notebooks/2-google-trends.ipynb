```python
# import ticker list from exploratory
tickers = pd.read_csv('../data/interim/unique_tickers.csv')
```

```python
tickers_list = tickers.columns.values.tolist()
```

```python
# remove extra ' from each ticker
tickers_list2 = [s.strip("'") for s in tickers_list]

# add 'stock to each of these Tickers'
final_tickers_list = [s + ' stock' for s in tickers_list2]
```

```python
# pytrends seems to time out around 90 requests. We are sending 4 kws in each request, so that's 360 kws per reqeust
# let's split into 10 lists of about 360
len(final_tickers_list)
```

```python
start = 0
stop = 360
tickers_dict = {}

for i in range(1, 11):
    if stop <= len(final_tickers_list) -360:
        print(i)
        list_name = 'tickers' + str(i)
        print(list_name)
        print(start, stop)
        lst =final_tickers_list[start:stop]
        tickers_dict[list_name] = lst
        print(list_name)
        print(tickers_dict)
        start +=361
        stop +=361
    else:
        print(i)
        list_name = 'tickers'+ str(i)
        print(start, stop)
        lst =final_tickers_list[start:stop]
        tickers_dict[list_name] = lst
        print(list_name)
        print(tickers_dict)
        start +=361
        stop +=300

    
```

```python
# pass each consecutive list, but with a wait time in between 
for key, value in tickers_dict.items():
    print("starting " + str(key))
    kw_list = value
    i= 0
    i2 = 1
    directory = "../data/interim/"
    key_ref = 'best stocks'
    trends = pd.DataFrame()
    
    while i < len(kw_list):
        l = kw_list[i:i+4]  
        l.append(key_ref)
        print(l)
        pytrends.build_payload(l, cat = 0, timeframe= '2021-01-01 2022-03-01', geo= 'US')
        add_trends = pytrends.interest_over_time()
        add_trends.drop(labels =['isPartial'], axis='columns', inplace=True)
        add_trends = add_trends.add_suffix('_' + str(i2))
        add_trends.reset_index(inplace=True)
        print(add_trends)

        if i == 0:
            print("i is 0")
            trends = add_trends
            print(trends)
        else:
            print("i is greater than 0")
            print("add trends is")
            print(add_trends)
            trends = pd.merge(trends, add_trends, how="left", left_on="date", right_on= "date")
            # print(trends)
        i+=4
        i2+=1
        time.sleep(120)
    trends_file_name = "gtrends" + str(key) + ".csv"
    print("saving to csv")
    trends.to_csv(directory + trends_file_name)
    time.sleep(200)
```
