```python
kw_dict= { 
 'technology-and-telecom': 'technology stocks',
 'consumer-goods': 'consumer goods stocks',
 'industrials': 'industrials stocks',
 'energy-materials-and-utilities': 'energy stocks',
 'investment-planning': 'investment planning',
 'real estate': 'real estate investing',
 'cryptocurrency': 'crytpo',
 'financials': 'financials stocks',
 'health-care': 'health care stocks',
 'cannabis': 'cannabis stocks',
 'markets': 'stock markets',
 'marijuana': 'cannabis stocks' ,
 'general 1' : 'best stocks',
 'general2' : 'stocks to buy'
}
```

```python
kw_dict
```

```python
# pass each consecutive list, but with a wait time in between 
i = 0

for key, value in kw_dict.items():
  
    print("topic is " + str(value))
    # add topic to a kw list
    kw = []
    kw.append(value)
    print(kw)

    # build request to google trends api
    print("building payload")
    pytrends.build_payload(kw, cat = 0, timeframe= '2021-01-01 2022-03-01', geo= 'US')
    
    # execute call to pytrends 
    print("sending to gtrends api")
    new_trends = pytrends.interest_over_time()
    
    # save as dataframe
    print("saving as data frame")
    add_trends = pd.DataFrame(new_trends)
    
    print(add_trends)
    
    # drop partial label
    print("dropping partial label")
    add_trends.drop(labels =['isPartial'], axis='columns', inplace=True)
    add_trends.reset_index(inplace=True)
    
    print("resetting index")
    
    print(add_trends)
    
    # for the first pull, set trends equal to add trends 
    if i == 0:
        print("i is 0")
        trends = add_trends
        print(trends)
    
    # for next pulls, merge data on to trends df
    else:
        print("i is greater than 0")
        print("trends is")
        print(trends)
        print("add trends is")
        print(add_trends)
        trends = pd.merge(trends, add_trends, how="left", left_on="date", right_on= "date")
        print("trends merged is ")
        print(trends)
    time.sleep(30)
    i+=1
#     trends_file_name = "gtrends" + str(key) + ".csv"
#     print("saving to csv")
#     trends.to_csv(directory + trends_file_name)
#     time.sleep(200)
```

### Deciding to fill in missing time series (this is weekly, and would prefer daily for an easier join)

```python
trends = pd.read_csv('../data/interim/gtrends.csv', parse_dates = True)
```

```python
trends.info()
```

```python
trends['date'] = pd.to_datetime(trends['date'])
```

```python
trends.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
trends_date_df = trends.set_index('date')
```

```python
trends_date_df
```

```python
trends_date_df_filled = trends_date_df.resample('1D').mean().ffill()
```

```python
trends_date_df_filled
```

```python
trends_df = trends_date_df_filled.reset_index()
```

```python
trends_df
```

```python
trends_melted = pd.melt(trends_df, id_vars=['date'], value_vars=trends_df.columns.to_list()[1:])
```

```python
trends_melted
```

```python
# reverse
rev_kw_dict = dict(zip(kw_dict.values(), kw_dict.keys()))
```

```python
trends_melted['bureau'] = trends_melted['variable'].map(rev_kw_dict)
```
