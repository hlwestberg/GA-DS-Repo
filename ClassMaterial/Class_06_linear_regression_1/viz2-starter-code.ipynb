### Subplots

In addition to exposing the figure and axes objects, the subplots will produce multiple axes objects on a figure.  These will be an array of axes objects if we enter a shape argument for the arangement of rows and columns in the subplots.

````python
###subplots 1 row 2 columns
# the ax here contains both axes below
fig, ax = plt.subplots(nrows = 1, ncols =2)
````

````python
#examine the ax object
# ax is an array of axes
ax
````

````python
#now plot to each
fig, ax = plt.subplots(1, 2);
ax[0].plot(x, f(x));
ax[1].hist(f(x));
````

````python
#control the figure size
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
#plot to the first axis
ax[0].plot(x, f(x));
ax[0].set_xlabel('x')
ax[0].set_title('f(x)')
ax[0].grid()
````

 Create a 1 column 3 row plot with:
 - a scatterplot of `balance` vs. `age`
 - an `age` boxplot
 - a bar plot of counts of the `marital` feature

````python
# to get the counts, you store this as a series, which has index and values. the index becomes the marital status and 
# then the values are the counts that we want to plot
married= vehicle['marital'].value_counts()
````

````python
fig, ax = plt.subplots(nrows = 3, ncols =1)
````

````python
fig, ax = plt.subplots(3,1 , figsize=(15, 5))
ax[0].scatter(vehicle['balance'], vehicle['age'])
ax[1].boxplot(vehicle['age'])
ax[2].bar(x=married.index, height=married.values)

# series has an index and values 
````
