#### Explore the dataset using visualizations

```python
# how does the popularity of football change over time?
plt.plot(sports['Year'], sports['Football'])
```

```python
# compare different sports over time
y = sports['Football']
y2 = sports['Volleyball']
y3= sports['Basketball']
x = sports['Year']

fig, ax = plt.subplots()

ax.plot(x, y)
ax.plot(x, y2)
ax.plot(x, y3)
plt.show()
```

```python
# avg pc by sport
sports2 = sports.set_index('Year')
# DataFrame.set_index(keys, drop=True, append=False, inplace=False, verify_integrity=False)[source]¶
```
